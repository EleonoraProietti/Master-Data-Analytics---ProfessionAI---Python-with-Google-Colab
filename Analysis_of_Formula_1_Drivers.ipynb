### Attribuzione punteggio

```python
# Punteggio in base alla posizione di arrivo

punteggio = {
    1: 10,
    2: 8,
    3: 6,
    4: 5,
    5: 4,
    6: 3,
    7: 2,
    8: 1
    }

# Dal nono posto non ottengono punteggio.
# Chi ha come posizione zero non Ã¨ arrivato al traguardo e anche lui non ha punteggio.
```

### 1. Analisi performance individuali dei piloti

```python
import csv as csv
import pandas as pd # Import pandas library

with open('formula1_data.csv') as (csv_file):
  csv_reader = csv.DictReader(csv_file)
  df = pd.read_csv('formula1_data.csv') # Load the CSV into a pandas DataFrame

# Riporta tre informazioni base sul pilota: punti totali, totale vittorie e totale podi ottenuti.

def performance_piloti (pilota):
  dati_pilota = df[df['Driver'] == pilota]

  punti_totali = 0
  vittorie = 0
  podi = 0

  for pos_str in dati_pilota['Position']:
    try:
      pos = int(pos_str)
      if pos in punteggio:
        punti_totali += punteggio[pos]
      if pos == 1:
        vittorie += 1
      if pos in [1, 2, 3]:
        podi += 1
    except ValueError:
      pass

  return [punti_totali, vittorie, podi]

performance_piloti('Hamilton')

# Abbiamo in output il punteggio totale del pilota, quante vittorie ha avuto nel campionato e il numero di podi complessivi.
```
